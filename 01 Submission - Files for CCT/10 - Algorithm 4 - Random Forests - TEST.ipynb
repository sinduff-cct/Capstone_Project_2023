```python
# Import libraries

import pandas as pd
import numpy as np

import seaborn as sns #visualisation
import matplotlib.pyplot as plt #visualisation

%matplotlib inline 
sns.set(color_codes=True)

# create tables
from tabulate import tabulate
```

### Import Test data file

```python
# read in the .csv file - of filtered student_info

data_df = pd.read_csv("test_data.csv")
```

```python
data_df.info()
```

### Algorithm 4 - Random Forest - Classifier

```python
# import libraries
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# divide the dataset into features (X) and the target variable (y)

# variable 1 = gender
X = data_df.drop(columns=['gender'])
y = data_df['gender']        

# variable 2 = studied_credits
X2 = data_df.drop(columns=['studied_credits'])
y2 = data_df['studied_credits']

# variable 3 = tenure
X3 = data_df.drop(columns=['tenure'])
y3 = data_df['tenure']     
```

```python
# split the data into training and test sets

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y2, test_size=0.2, random_state=0)
X_train3, X_test3, y_train3, y_test3 = train_test_split(X3, y3, test_size=0.2, random_state=0)
```

```python
# create Random Forest with entrophy and 200 'trees' in the forest

# Test 2@ 200 'trees'
classifier_rf = RandomForestClassifier(n_estimators=200, criterion='entropy', random_state=0)
```

### Hyperparm tuning - GridSearch CV

```python
# define the parameter values that should be searched
# instance, max_depth parameter can be any integer from 1 to 20
# max_features can be auto, sqrt, log2 or None

param_grid_rf = {'max_depth': list(range(1, 21)), 
                 'max_features': ['auto', 'log2', None]
                }
```

```python
# instantiate the grid for data search
grid_rf = GridSearchCV(classifier_rf, param_grid_rf, cv=10, scoring='accuracy')

# fit the grid with data
grid_rf.fit(X_train, y_train)
```

```python
# show the best model and retrain classifier 

print("Best Model Accuracy:",grid_rf.best_score_)
print("Best Hyperparameters:",grid_rf.best_params_)
```

```python
# train the model using the training sets

classifier_rf = RandomForestClassifier(n_estimators=100, criterion='entropy', 
                                       max_depth=grid_rf.best_params_['max_depth'], 
                                       max_features=grid_rf.best_params_['max_features'], 
                                       random_state=0)
```

```python
classifier_rf.fit(X_train, y_train)
```

```python
# make predictions on the test set

y_pred_rf = classifier_rf.predict(X_test)
```

```python
# calculate model accuracy

print("Accuracy for the model is:", accuracy_score(y_test, y_pred_rf))
```

### Fit and tune with model for variable 2 - studied_credits¶
